## example 1. pima indian diabetes dataset
First let's load it in

```python
import pandas as pd
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
#cols = ['pregnant', 'glucose', 'bp', 'skinFold', 'insulin', 'bmi', 'pedigree', 'age', 'diabetes']
diabetes = pd.read_csv('../data/diabetes.csv')
diabetes
```

### Normalize
And we will normalize everything using the `MinMaxScaler`

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
diabetes[columns] =  scaler.fit_transform(diabetes[columns] )
diabetes
```

### Divide into training and testing

```python
from sklearn.model_selection import train_test_split
diabetes_train, diabetes_test = train_test_split(diabetes, test_size = 0.2)
len(diabetes_test)
```

### now build classifier

Let's use a kNN classifier with k=1

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(diabetes[columns[:-1]], diabetes[columns[-1]])
```

### Finally, let's make predictions and compute the accuracy:

```python
### now make predictions and compute accuracy:
predictions = knn.predict(diabetes_test[columns[:-1]])
from sklearn.metrics import accuracy_score
accuracy_score(diabetes_test['Outcome'], predictions)
```

### WOW! 100% accurate on 154 samples of our test set!

Let's try ...
#### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(diabetes[columns[:-1]], diabetes[columns[-1]])
bayes_predictions = clf.predict(diabetes_test[columns[:-1]])

accuracy_score(diabetes_test['Outcome'], bayes_predictions)
```

Hmm. not as good as kNN. But that 100% should through up some red flags. I'm going to add some vertical white space so you have time to ponder this ....

# New features

### Maybe we can improve performance by adding new features

Here is a simple example, but a bit convoluted.

Suppose instead of BMI our initial dataset had height and weight. This is the *let's pretend* bit that is not part of what I am trying to show.  

Let's create that dataset.

```python
import pandas as pd
import numpy as np
modified_columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'Height', 'Weight', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
#cols = ['pregnant', 'glucose', 'bp', 'skinFold', 'insulin', 'bmi', 'pedigree', 'age', 'diabetes']
modified_diabetes = pd.read_csv('../data/diabetes.csv')
modified_diabetes['Height'] = np.random.randint(55, 70, size=len(modified_diabetes))
modified_diabetes['Weight'] = (modified_diabetes['Height'] * modified_diabetes['Height']) * modified_diabetes['BMI'] / 703
#modified_diabetes = modified_diabetes.drop('BMI', 1)
modified_diabetes = modified_diabetes[modified_columns]
modified_diabetes
```

## now normalize

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalized_modified_diabetes = modified_diabetes[modified_columns] 
normalized_modified_diabetes[modified_columns] =  scaler.fit_transform(normalized_modified_diabetes[modified_columns] )
normalized_modified_diabetes
```

Imagine this were our original dataset. 

### Now look at  the ANOVA scores for the features


```python
from sklearn.feature_selection import SelectKBest
selector = SelectKBest(k=6)
selector.fit(normalized_modified_diabetes[modified_columns[:-1]], normalized_modified_diabetes['Outcome'])
scores = {normalized_modified_diabetes.columns[i]:selector.scores_[i] for i in range(len(normalized_modified_diabetes.columns)-1)}
sorted_features = sorted(scores,key=scores.get, reverse=True)
for feature in sorted_features:
    print('Feature %s has value %f'%(feature,scores[feature]))
    
```

So height has little relationship with diabetes and weight has only a value of 65. We might think

#### Is there any new feature we can add that will improve performance?

If our original dataset had height and weight we might consider using them to compute BMI and we can check whether BMI is a good predictor. So let's add a new column for BMI

```python

modified_diabetes['bmi'] = modified_diabetes['Weight'] * 703 /  (modified_diabetes['Height'] * modified_diabetes['Height'])
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'bmi', 'Height', 'Weight', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
scaler = MinMaxScaler()
bmi_modified_diabetes = modified_diabetes[columns] 
bmi_modified_diabetes[modified_columns] =  scaler.fit_transform(normalized_modified_diabetes[modified_columns] )
selector = SelectKBest( k=6)
selector.fit(bmi_modified_diabetes[columns[:-1]], bmi_modified_diabetes['Outcome'])
scores = {bmi_modified_diabetes.columns[i]:selector.scores_[i] for i in range(len(bmi_modified_diabetes.columns)-1)}
sorted_features = sorted(scores,key=scores.get, reverse=True)
for feature in sorted_features:
    print('Feature %s has value %f'%(feature,scores[feature]))
```

### So our new feature, BMI looks like it will improve performance


# Final thoughts on that topic

A great candidate for feature reduction is text learning. In the Enron email dataset we had 38,757 different features. That is a lot of dimensions. We could reduce that to a more manageable size and perhaps not lose accuracy.

Adding features might also be useful in text learning.  For example, consider identifying whether in Congress is female or male. Knowing the frequency of *said* or *we* probably won't help us much but two word sequences (called word bigrams) have been shown to be good predictors. I'll give you 2 sets of bigrams. Can you decide which set is common to female speakers and which to male?

* *we protect*
* *we share* 
* *we honor*
* *our community*
* *our elderly*

And the other set:

* *we gave*
* *we said* 
* *we took*
* *our enemy*
* *our side*

So adding these bigrams as features would improve performance
